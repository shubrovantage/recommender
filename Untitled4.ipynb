```python
import pandas as pd
import time
from flask import Flask, jsonify, request


app = Flask(__name__)
start=time.time()

# 455
active_455 = pd.read_csv('/Users/vantagecircle/Downloads/455/active_employees.csv')
feed_455 = pd.read_csv('/Users/vantagecircle/Downloads/455/feed_data.csv')


# 1159
active_1159 = pd.read_csv('/Users/vantagecircle/Downloads/1159/active_employees.csv')
feed_1159 = pd.read_csv('/Users/vantagecircle/Downloads/1159/feed_data.csv')


#1065
active_1065 = pd.read_csv('/Users/vantagecircle/Downloads/1065/active_employees.csv')
feed_1065 = pd.read_csv('/Users/vantagecircle/Downloads/1065/feed_data.csv')


# 355
active_355 = pd.read_csv('/Users/vantagecircle/Downloads/active_employees.csv')
feed_355 = pd.read_csv('/Users/vantagecircle/Downloads/feed_data.csv')

# 63
active_63 = pd.read_csv('/Users/vantagecircle/Downloads/active_employees_wipro.csv')
feed_63 = pd.read_csv('/Users/vantagecircle/Downloads/feeddata_wipro.csv')

# assigning company columns
active_455['company_id'] = 455
active_1159['company_id'] = 1159
active_1065['company_id'] = 1065
active_355['company_id'] = 355
active_63['company_id'] = 63

df_active_employees = pd.concat([active_455,active_1159,active_1065,active_355,active_63],ignore_index=True)
df_feed_data = pd.concat([feed_455,feed_1159,feed_1065,feed_63,feed_355],ignore_index=True)
#df_feed_likes = pd.concat([likes_1065,likes_1159,likes_355,likes_455,likes_63],ignore_index=True)
#df_feed_comments = pd.concat([comments_1065,comments_1159,comments_355,comments_455,comments_63],ignore_index=True)

df_active_employees.rename(columns={'id':'employee_id'},inplace=True)
df_active_employees['department'] = df_active_employees['department'].fillna('Other')

columns = ['id', 'feed_type', 'sender_id', 'receiver_ids', 'created_at']
df = pd.DataFrame(df_feed_data, columns=columns)

# Split rows with multiple receiver IDs
df['receiver_ids'] = df['receiver_ids'].str.split(',')
df = df.explode('receiver_ids', ignore_index=True)

# Reset index and create a new DataFrame with the same columns
new_df = df.reset_index(drop=True)

# Convert 'created_at' column to datetime
new_df['created_at'] = pd.to_datetime(new_df['created_at'])

# Drop rows with NaN values in 'sender_id' column
new_df.dropna(subset=['sender_id'], inplace=True)

# Convert 'sender_id' column to integer
new_df['sender_id'] = new_df['sender_id'].astype(int)

new_df.rename(columns={'sender_id':'sender_employee_id','receiver_ids':'receiver_employee_id'},inplace=True)
new_df.dropna(subset=['receiver_employee_id','sender_employee_id'],inplace=True)
new_df['receiver_employee_id'] = new_df['receiver_employee_id'].astype(int)

# Assuming you have the 'employee_id' column in both DataFrames
merged_df = new_df.merge(df_active_employees, left_on='sender_employee_id', right_on='employee_id', how='left')
merged_df.rename(columns={'department': 'sender_department'}, inplace=True)
merged_df.drop(columns='employee_id', inplace=True)
merged_df = merged_df.merge(df_active_employees, left_on='receiver_employee_id', right_on='employee_id', how='left')
merged_df.rename(columns={'department': 'receiver_department'}, inplace=True)
merged_df.drop(columns='employee_id', inplace=True)
merged_df.drop(columns=['name_x','name_y'],inplace=True)
merged_df.rename(columns={'receiver_department':'receiver_dept'},inplace=True)
merged_df.rename(columns={'sender_department':'sender_dept'},inplace=True)
merged_df.rename(columns={'company_id_y':'receiver_company'},inplace=True)
merged_df.rename(columns={'company_id_x':'sender_company'},inplace=True)

new_df=merged_df
new_df['created_at']=pd.to_datetime(new_df['created_at'])




# function to assign feed_type strength
def feed_type_strength(df):
  feed_type_strength = {'award':2,'badge':1}
  df['feedStrength']=df['feed_type'].apply(lambda x: feed_type_strength[x])
  return df
feed_type_strength(new_df)

start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-06-30')
# Create a boolean mask to filter the DataFrame
date_mask = (new_df['created_at'] >= start_date) & (new_df['created_at'] <= end_date)
# Apply the mask to filter the DataFrame
fil_df = new_df[date_mask]

grouped_df=fil_df.groupby(['receiver_dept','receiver_company','receiver_employee_id'])['feedStrength'].sum().reset_index()


def get_team_recommendation(chosen_employee_id):
    employee=int(chosen_employee_id)

    department_name = df_active_employees[df_active_employees['employee_id'] == employee]['department'].values[0]
    company_id = int(df_active_employees[df_active_employees['employee_id']==employee]['company_id'])
    dept_df=grouped_df[(grouped_df['receiver_dept']==department_name) & (grouped_df['receiver_company']==company_id)]
    dept_df=dept_df.sort_values(by='feedStrength',ascending=False)
    dept_df=dept_df[dept_df['receiver_employee_id']!=employee]
    dept_df=dept_df.head(10)

    other_df = fil_df[(fil_df['sender_employee_id'] == employee) & (fil_df['receiver_dept'] != department_name) & (fil_df['receiver_company'] == company_id)]
    other_grouped = other_df.groupby(['receiver_employee_id', 'receiver_company', 'receiver_dept'])['feedStrength'].sum().reset_index()
    other_grouped = other_grouped.sort_values(by='feedStrength', ascending=False).head(10)
    return dept_df

end=time.time()
print(end-start)

@app.route('/get_team_recommendations_api', methods=['GET'])
def get_team_recommendations_api():
    chosen_employee_id = request.args.get('chosen_employee_id')
    print("Here: ", chosen_employee_id)
    dept_df = get_team_recommendation(chosen_employee_id)
    other_grouped = get_team_recommendation(chosen_employee_id)

    dept_data = dept_df.to_dict(orient='records')
    other_data = other_grouped.to_dict(orient='records')

    return jsonify({'dept_df': dept_data,'other_df':other_data})

if __name__ == '__main__':
    app.run(debug=True)
```
